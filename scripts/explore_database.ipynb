# Explore NewsAgent Database

This notebook demonstrates how to explore and inspect database content saved by NewsAgent.

Topics:
1. Connect to database
2. View database statistics
3. Query stored queries (QueryRequest)
4. Query stored responses (QueryResponse)
5. Query stored items (SearchItem)
6. Analyze data
7. Export results

## Step 1: Setup and Import

```python
import sys
from pathlib import Path
import json

# Add parent directory to path
parent_dir = str(Path.cwd().parent.resolve())
if parent_dir not in sys.path:
    sys.path.insert(0, parent_dir)

# Import NewsAgent modules
from src.database import SQLite3Backend, DatabaseManager
from src.scheduler.scheduler_settings import SchedulerSettings

print("✓ Imports successful")
```

## Step 2: Connect to Database

```python
print("=" * 70)
print("CONNECTING TO DATABASE")
print("=" * 70)
print()

# Get settings
settings = SchedulerSettings.initialize()
db_path = settings.env_vars.database_path

print(f"Database path: {db_path}")
print()

# Connect to database
db_backend = SQLite3Backend(db_path=db_path)
db_manager = DatabaseManager(db_backend)
db_manager.connect()

print("✓ Database connected successfully")
print()
```

## Step 3: View Database Statistics

```python
print("=" * 70)
print("DATABASE STATISTICS")
print("=" * 70)
print()

db_manager.print_stats()
print()
```

## Step 4: Explore Stored Queries (QueryRequest)

View all QueryRequest objects that were executed.

```python
print("=" * 70)
print("STORED QUERIES (QueryRequest)")
print("=" * 70)
print()

# Get all queries
queries = db_manager.list_queries(limit=100)
print(f"Total queries stored: {len(queries)}")
print()

if queries:
    # Show details of each query
    for i, query in enumerate(queries[:5], 1):
        print(f"[{i}] Query ID: {query.query_id}")
        print(f"    Fields: {', '.join(query.query_fields)}")
        print(f"    Topics: {', '.join(query.query_topics)}")
        print(f"    Agents: {', '.join(query.source_agents)}")
        print(f"    Days back: {query.days_back}")
        print(f"    Max results: {query.max_results}")
        print(f"    AI summary: {query.include_ai_summary}")
        print(f"    Raw response: {query.include_raw_response}")
        print(f"    Timestamp: {query.timestamp}")
        print()
else:
    print("❌ No queries found in database")
    print("   Run demo_scheduler_bocha.ipynb first to save queries")
```

## Step 5: Explore Stored Responses (QueryResponse)

View all QueryResponse objects that were executed.

```python
print("=" * 70)
print("STORED RESPONSES (QueryResponse)")
print("=" * 70)
print()

# Get all responses
responses = db_manager.list_responses(limit=100)
print(f"Total responses stored: {len(responses)}")
print()

if responses:
    # Show details of each response
    for i, response in enumerate(responses[:5], 1):
        print(f"[{i}] Response ID: {response.response_id}")
        print(f"    Agent: {response.agent_name}")
        print(f"    Query ID: {response.query_id}")
        print(f"    Success: {response.success}")
        print(f"    Status: {response.status}")
        print(f"    Items: {len(response.items)}")
        print(f"    Total estimated: {response.total_estimated}")
        print(f"    Execution time: {response.execution_time_ms}ms")
        print(f"    Timestamp: {response.timestamp}")
        
        # Show error if present
        if response.error_message:
            print(f"    Error: {response.error_message}")
        
        print()
else:
    print("❌ No responses found in database")
    print("   Run demo_scheduler_bocha.ipynb first to save responses")
```

## Step 6: Explore Stored Items (SearchItem)

View all SearchItem objects that were parsed from API responses.

```python
print("=" * 70)
print("STORED ITEMS (SearchItem)")
print("=" * 70)
print()

# Get recent items
items = db_manager.get_recent_items(days=30, limit=100)
print(f"Total items (last 30 days): {len(items)}")
print()

if items:
    # Show first 5 items
    for i, item in enumerate(items[:5], 1):
        print(f"[{i}] {item.title}")
        print(f"    ID: {item.id}")
        print(f"    Source: {item.source_name} ({item.source_type})")
        print(f"    URL: {item.source_url}")
        print(f"    Content: {item.content[:100]}...")
        print(f"    Published: {item.timestamp}")
        
        # Show metadata if present
        if item.metadata:
            print(f"    Metadata:")
            for key, value in item.metadata.items():
                if isinstance(value, str):
                    preview = value[:60] + "..." if len(value) > 60 else value
                else:
                    preview = str(value)[:60]
                print(f"      {key}: {preview}")
        
        print()
else:
    print("❌ No items found in database")
    print("   Run demo_scheduler_bocha.ipynb first to save items")
```

## Step 7: Search by Agent

```python
print("=" * 70)
print("SEARCH ITEMS BY AGENT")
print("=" * 70)
print()

# Search for BOCHA items
bocha_items = db_manager.search_items(source_type="BOCHA", limit=50)
print(f"Items from BOCHA: {len(bocha_items)}")

# Count by source
source_types = {}
all_items = db_manager.get_recent_items(days=30, limit=1000)
for item in all_items:
    source = item.source_type
    source_types[source] = source_types.get(source, 0) + 1

print("\nItems by source type:")
for source, count in sorted(source_types.items(), key=lambda x: x[1], reverse=True):
    print(f"  {source}: {count}")
print()
```

## Step 8: Query by Agent Name

```python
print("=" * 70)
print("QUERY RESPONSES BY AGENT")
print("=" * 70)
print()

# Get BOCHA responses
bocha_responses = db_manager.list_responses(agent_name="BOCHA", limit=50)
print(f"Responses from BOCHA: {len(bocha_responses)}")
print()

if bocha_responses:
    # Statistics
    successful = sum(1 for r in bocha_responses if r.success)
    failed = sum(1 for r in bocha_responses if not r.success)
    total_items = sum(len(r.items) for r in bocha_responses)
    avg_items = total_items / len(bocha_responses) if bocha_responses else 0
    
    print("BOCHA Response Statistics:")
    print(f"  Total responses: {len(bocha_responses)}")
    print(f"  Successful: {successful}")
    print(f"  Failed: {failed}")
    print(f"  Total items retrieved: {total_items}")
    print(f"  Average items per response: {avg_items:.1f}")
else:
    print("❌ No BOCHA responses found")
```

## Step 9: Load Specific Query and Response

```python
print("=" * 70)
print("LOAD SPECIFIC QUERY AND RESPONSE")
print("=" * 70)
print()

# Get first query and response
all_queries = db_manager.list_queries(limit=1)

if all_queries:
    query = all_queries[0]
    print(f"Query ID: {query.query_id}")
    print()
    
    # Try to load from database
    loaded_query = db_manager.load_query(query.query_id)
    print("Loaded Query:")
    print(f"  Fields: {loaded_query.query_fields}")
    print(f"  Topics: {loaded_query.query_topics}")
    print(f"  Agents: {loaded_query.source_agents}")
    print(f"  Days back: {loaded_query.days_back}")
    print(f"  Max results: {loaded_query.max_results}")
    print()
    
    # Find responses for this query
    all_responses = db_manager.list_responses(limit=100)
    matching_responses = [r for r in all_responses if r.query_id == query.query_id]
    
    print(f"Responses for this query: {len(matching_responses)}")
    print()
    
    if matching_responses:
        response = matching_responses[0]
        print(f"Response ID: {response.response_id}")
        print(f"  Agent: {response.agent_name}")
        print(f"  Success: {response.success}")
        print(f"  Status: {response.status}")
        print(f"  Items: {len(response.items)}")
        print()
        
        # Show items from this response
        if response.items:
            print(f"Items from this response ({len(response.items)} total):")
            for i, item in enumerate(response.items[:3], 1):
                print(f"  [{i}] {item.title}")
                print(f"      Source: {item.source_name}")
                print(f"      URL: {item.source_url}")
else:
    print("❌ No queries in database")
```

## Step 10: Verify Request-Response Relationship

```python
print("=" * 70)
print("VERIFY REQUEST-RESPONSE RELATIONSHIP")
print("=" * 70)
print()

# Get all queries and responses
all_queries = db_manager.list_queries(limit=100)
all_responses = db_manager.list_responses(limit=100)

print(f"Total queries in database: {len(all_queries)}")
print(f"Total responses in database: {len(all_responses)}")
print()

if all_queries and all_responses:
    # Create mapping
    query_response_map = {}
    for response in all_responses:
        query_id = response.query_id
        if query_id not in query_response_map:
            query_response_map[query_id] = []
        query_response_map[query_id].append(response)
    
    # Show relationship
    print("Query → Response Mapping:")
    print()
    for query in all_queries[:5]:
        responses = query_response_map.get(query.query_id, [])
        print(f"Query: {query.query_id[:8]}...")
        print(f"  Fields: {', '.join(query.query_fields[:2])}...")
        print(f"  Responses: {len(responses)}")
        
        for resp in responses:
            print(f"    - {resp.agent_name}: {len(resp.items)} items (success={resp.success})")
        print()
    
    # Summary
    print("Summary:")
    print(f"  Queries with responses: {len(query_response_map)}")
    print(f"  Queries without responses: {len([q for q in all_queries if q.query_id not in query_response_map])}")
    print(f"  Total request-response pairs: {len(all_responses)}")
else:
    print("❌ Not enough data in database to show relationship")
```

## Step 11: Export Data to JSON

```python
print("=" * 70)
print("EXPORT DATA TO JSON")
print("=" * 70)
print()

import os
from datetime import datetime

# Prepare export data
all_queries = db_manager.list_queries(limit=100)
all_responses = db_manager.list_responses(limit=100)
all_items = db_manager.get_recent_items(days=30, limit=1000)

export_data = {
    "metadata": {
        "export_date": datetime.now().isoformat(),
        "database_path": str(db_path),
        "total_queries": len(all_queries),
        "total_responses": len(all_responses),
        "total_items": len(all_items),
    },
    "queries": [
        {
            "query_id": q.query_id,
            "fields": q.query_fields,
            "topics": q.query_topics,
            "agents": q.source_agents,
            "days_back": q.days_back,
            "max_results": q.max_results,
            "timestamp": q.timestamp.isoformat(),
        }
        for q in all_queries[:10]
    ],
    "responses": [
        {
            "response_id": r.response_id,
            "agent_name": r.agent_name,
            "query_id": r.query_id,
            "success": r.success,
            "status": r.status,
            "items_count": len(r.items),
            "total_estimated": r.total_estimated,
            "execution_time_ms": r.execution_time_ms,
            "timestamp": r.timestamp.isoformat(),
        }
        for r in all_responses[:10]
    ],
    "sample_items": [
        {
            "id": item.id,
            "title": item.title,
            "source_name": item.source_name,
            "source_type": item.source_type,
            "source_url": item.source_url,
            "content": item.content[:200] + "..." if len(item.content) > 200 else item.content,
            "timestamp": item.timestamp.isoformat(),
        }
        for item in all_items[:10]
    ],
}

# Save to file
export_file = "data/database_export.json"
os.makedirs(os.path.dirname(export_file), exist_ok=True)

with open(export_file, 'w', encoding='utf-8') as f:
    json.dump(export_data, f, ensure_ascii=False, indent=2)

file_size = os.path.getsize(export_file)
print(f"✓ Data exported to: {export_file}")
print(f"  File size: {file_size:,} bytes")
print()
print("Export summary:")
print(f"  Queries: {len(export_data['queries'])}")
print(f"  Responses: {len(export_data['responses'])}")
print(f"  Sample items: {len(export_data['sample_items'])}")
```
